1 - Carregando um arquivo excel.

```python
import pandas as pd

arquivo = pd.read_excel("cursinho.xlsx")
display(arquivo)
```

2 - Total de presenças na semana.

```python
presentes = arquivo["Alunos presentes"].sum()
print(f"Total de prensenças na semana: {presentes}")
```

3 - Total de faltas na semana.

```python
faltas = arquivo["Alunos com falta"].sum()
print(f"Total de faltas na semana: {faltas}")
```

4 - Os 3 dias e matérias com mais faltas na semana.

```python
mais_faltas = arquivo.sort_values(by='Alunos com falta', ascending=False)

top_3 = mais_faltas.iloc[:3, :]

print(top_3)
```

5- Os 3 dias e matérias com mais presenças esta semana.

```python
mais_presen = arquivo.sort_values(by='Alunos com falta')

top_3 = mais_presen.iloc[:3, :]

print(top_3)
```
